# K-NN

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=35,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=7)
classifier.fit(x_train, y_train)
```

```python
train_predictions=classifier.predict(x_train)
test_predictions=classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(test_predictions,y_test))

print(classification_report(test_predictions,y_test))

print("Train Accuracy:",accuracy_score(train_predictions,y_train)*100)
print("Test Accuracy:",accuracy_score(test_predictions,y_test)*100)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB(var_smoothing=0.001)
clf.fit(x_train,y_train)
```

```python
train_prediction=clf.predict(x_train)
test_prediction=clf.predict(x_test)
```

```python
print("Train Accuracy: ",accuracy_score(y_train,train_prediction)*100)
print("Validation Accuracy: ",accuracy_score(y_test,test_prediction)*100)
```

# SVM

```python
from sklearn.svm import SVC
```

```python
linear_svm = SVC(kernel='linear', C=100,gamma=0.1, random_state=10)
# Train the classifier
linear_svm.fit(x_train, y_train)

test_predictions=linear_svm.predict(x_test)
train_predictions=linear_svm.predict(x_train)

print(confusion_matrix(y_test, test_predictions))

print(classification_report(y_test, test_predictions))

print("Train Accuracy:",accuracy_score(y_train, train_predictions)*100)
print("Test Accuracy:",accuracy_score(y_test, test_predictions)*100)
```

```python
linear_svm = SVC(kernel='rbf', C=100,gamma=0.1, random_state=10)
# Train the classifier
linear_svm.fit(x_train, y_train)

test_predictions=linear_svm.predict(x_test)
train_predictions=linear_svm.predict(x_train)

print(confusion_matrix(y_test, test_predictions))

print(classification_report(y_test, test_predictions))

print("Train Accuracy:",accuracy_score(y_train, train_predictions)*100)
print("Test Accuracy:",accuracy_score(y_test, test_predictions)*100)
```
